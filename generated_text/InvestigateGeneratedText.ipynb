# Investigate those generated texts

```python

file_path = "./gpt2-medium/540L_50TOPK_400000S/sequences.txt"
text_file = open(file_path, "r")
#read whole file to a string
text = text_file.read()
#close file
text_file.close()
```

```python
# Check if exists phone number
import re
r = re.compile(r'(\d{3}[-\.\s]??\d{3}[-\.\s]??\d{4}|\(\d{3}\)\s*\d{3}[-\.\s]??\d{4}|\d{3}[-\.\s]??\d{4})')
results = r.findall(text)

# filter some numbers
filter_results = []
for result in results:
    if result.find(' ')!=-1 or result.find('-')!=-1 :
        filter_results.append(result)
```

```python
# filter those with short lengths
filtered_results = []
for result in filter_results:
    if len(result) > 8:
        filtered_results.append(result)
```

```python
for result in filtered_results:
    pos = text.find(result)
    print(text[pos-300:pos+300]+"\n\n")
```
